```python
import librosa
import noisereduce as nr
import matplotlib.pyplot as plt
from IPython.display import Audio 
import numpy as np 

# Load audio file
audio_file = './dataset/High20m_50m_#3.wav'  # Replace with your audio file path
y, fs = librosa.load(audio_file)  # , duration=20)

# Reduce noise
reduced_noise = nr.reduce_noise(y, sr=fs)

t = np.linspace(0, len(y) / fs, len(y))
# Plot original and reduced noise signals
fig, ax = plt.subplots(2, 1, figsize=(15,8))
ax[0].set_title("Original signal")
ax[0].plot(t,y)
ax[1].set_title("Reduced noise signal")
ax[1].plot(t,reduced_noise)
plt.show()
```
